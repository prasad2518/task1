```python
import numpy as np
import matplotlib.pyplot as plt

def sum_(x):
    s=0
    for i in x:
        s+=i
    return s
def length(x):
    n=0
    for i in x:
        n+=1
    return n
def mean_(x):
    return (sum_(x)/length(x))

#reading inputs
a=[float(x) for x in input("Enter x values : ").split(" ")]
b=[float(x) for x in input("Enter y values : ").split(" ")]
x=np.array(a)
y=np.array(b)

n=length(x)
sx=sum_(x)
sy=sum_(y)
sxy=sum_(x*y)
sx2=sum_(x*x)

a=((sy*sx2)-(sx*sxy))/((n*sx2)-(sx*sx))
b=(sy-(n*a))/sx
a=round(a,4)
b=round(b,4)

print("Slope is : ",b)
print("Intercept is : ",a)

#Best Fit Equation is:
print("Best Fit Equation is:")
print("y={}+{}*x".format(a,b))

#calculating y_hat
y_cap=b*x+a
print(y_cap)

#plots
plt.plot(x,y,'ro-')
plt.plot(x,y_cap,'bo-')
plt.show()

#sse
sse=sum((y-y_cap)**2)
print("sse is : ",sse)

#sst
sst=sum((y-mean_(y))**2)
print("sst is : ",sst)

#ssr
ssr=sst-sse
print("ssr is : ",ssr)

#calculating r2
r2=ssr/sst
print("r^2 is : ",r2)

#goodness of fit
if(r2>0.90):
  print("Good Fit")
else:
  print("Not Good Fit")

```
